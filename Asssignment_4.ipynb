```python
def get_recession_start():
    gdp_dt = pd.read_excel('gdplev.xls', skiprows = 219)
    gdp_dt = gdp_dt[['1999q4',9926.1]]
    gdp_dt.columns = ['Quarter', 'GDP']
    gdp_dt['GDP'] = pd.to_numeric(gdp_dt['GDP'])
    quarters = []
    for i in range(len(gdp_dt)-2):
        if (gdp_dt.iloc[i][1] > gdp_dt.iloc[i+1][1]) & (gdp_dt.iloc[i+1][1] > gdp_dt.iloc[i+2][1]):
            quarters.append(gdp_dt.iloc[i][0])
    return quarters[0]
        
get_recession_start()
    
```

```python
def get_recession_end():
    dt =  pd.read_excel('gdplev.xls', skiprows = 219)
    dt = dt[['1999q4',9926.1]]
    dt.columns = ['Quarter', 'GDP']
    dt['GDP'] = pd.to_numeric(dt['GDP'])
    st = get_recession_start()
    st_index = dt[dt['Quarter'] == st].index.tolist()[0]
    dt = dt.iloc[st_index:]
    for i in range(len(dt)):
        if (dt.iloc[i-2][1]< dt.iloc[i-1][1]) & (dt.iloc[i-1][1]< dt.iloc[i][1]):
            return dt.iloc[i][0]
    
get_recession_end()
    
    
```

```python
def get_recession_bottom():
    dt = pd.read_excel('gdplev.xls', skiprows = 219)
    dt = dt[['1999q4',9926.1]]
    dt.columns = ['Quarter', 'GDP']
    dt['GDP'] = pd.to_numeric(dt['GDP'])
    start = get_recession_start()
    start_index = dt[dt['Quarter'] == start].index.tolist()[0]
    end = get_recession_end()
    end_index = dt[dt['Quarter'] == end].index.tolist()[0]
    dt = dt.iloc[start_index:end_index + 1]
    bottom = dt['GDP'].min()
    bottom_index = dt[dt['GDP'] == bottom].index.tolist()[0]-start_index
    return dt.iloc[bottom_index]['Quarter']

get_recession_bottom()
```

```python
def convert_housing_data_to_quarters():
    all_homes = pd.read_csv('City_Zhvi_AllHomes.csv')
    all_homes['State'].replace(states, inplace = True)
    all_homes = all_homes.set_index(['State', 'RegionName'])
    all_homes = all_homes.iloc[:, 49:250]
    
    def quarters(col):
        if col.endswith(('01','02','03')):
            s = col[:4]+'q1'
        elif col.endswith(('04','05','06')):
            s = col[:4]+'q2'
        elif col.endswith(('07','08','09')):
            s = col[:4]+'q3'
        else:
            s = col[:4]+'q4'
            
        return s
    
    housing = all_homes.groupby(quarters, axis = 1).mean()
    housing = housing.sort_index()
    return housing


convert_housing_data_to_quarters()    
```

```python
def run_ttest():
  """
  First creates new data showing the decline or growth of housing prices
  between the recession start and the recession bottom. Then runs a ttest
  comparing the university town values to the non-university towns values, 
  return whether the alternative hypothesis (that the two groups are the same)
  is true or not as well as the p-value of the confidence. 
  
  Return the tuple (different, p, better) where different=True if the t-test is
  True at a p<0.01 (we reject the null hypothesis), or different=False if 
  otherwise (we cannot reject the null hypothesis). The variable p should
  be equal to the exact p value returned from scipy.stats.ttest_ind(). The
  value for better should be either "university town" or "non-university town"
  depending on which has a lower mean price ratio (which is equivilent to a
  reduced market loss).
  """
  # 選出衰退起始的前一季與衰退谷底當季的房價 ------------
  housing_data_quarters = convert_housing_data_to_quarters()
  recession_start = get_recession_start()
  recession_bottom = get_recession_bottom()
  quarters = housing_data_quarters.columns
  recession_start_idx = list(quarters).index(recession_start)
  before_recession = quarters[:recession_start_idx][-1]
  new_data = housing_data_quarters.loc[:, [before_recession, recession_bottom]].reset_index()
  # 判斷房價資料中的區域是否為大學城 ------------
  university_towns = get_list_of_university_towns()#.drop('State', axis=1)
  new_data = pd.merge(new_data, university_towns, how='left', left_on = ['State', 'RegionName'], right_on = ['State', 'Region Name'])
  new_data = new_data.assign(is_university_town = [True if isinstance(x, str) else False for x in new_data['State'].values])
  new_data = new_data.drop('State', axis=1)
  # 計算 price_ratio ------------
  new_data = new_data.assign(price_ratio = new_data[before_recession] / new_data[recession_bottom])
  # 執行 T 檢定 ------------
  university_towns = new_data[new_data['is_university_town']]['price_ratio'].values
  non_university_towns = new_data[new_data['is_university_town'] == False]['price_ratio'].values
  ttest_res = ttest_ind(university_towns, non_university_towns, nan_policy='omit')
  # 回傳答案 ------------
  reject_null_hypothesis = None
  lower_mean_price_ratio = None
  if ttest_res[1] < 0.01:
    reject_null_hypothesis = True
  else:
    reject_null_hypothesis = False
  if ttest_res[0] < 0:
    lower_mean_price_ratio = "university town"
  else:
    lower_mean_price_ratio = "non-university town"
  return reject_null_hypothesis, ttest_res[1], lower_mean_price_ratio
run_ttest()
```
